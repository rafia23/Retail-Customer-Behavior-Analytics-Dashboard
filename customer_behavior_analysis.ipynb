```python
import pandas as pd
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
#Replacing missing review ratings with the average rating of their own category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
#creating a column age_group
labels = [ 'Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels) #Equal number of people per group
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# creating column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df.head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
#checking if these two columns contain same value
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#removing promo code column (aixs=1)
df = df.drop('promo_code_used', axis=1)
```
